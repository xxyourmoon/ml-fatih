## 3. Import CSV ke Pandas DataFrame

Sesuai materi: **Importing Data to Pandas** menggunakan `pd.read_csv()`

```python
path = 'epl_goalscorer_2021.csv'
df = pd.read_csv(path)

print(f'Jumlah baris  : {df.shape[0]}')
print(f'Jumlah kolom  : {df.shape[1]}')
df
```

## 4. Melihat Data Awal dan Akhir

Sesuai materi: **`head()`** untuk baris pertama, **`tail()`** untuk baris terakhir.

```python
# 3 baris pertama
print('=== df.head(3) ===')
df.head(3)
```

```python
# 7 baris terakhir
print('=== df.tail(7) ===')
df.tail(7)
```

## 5. Inspeksi Tipe Data Setiap Kolom

Sesuai materi: **`df.dtypes`** untuk melihat tipe data tiap kolom.

```python
print('=== Tipe Data Setiap Kolom ===')
print(df.dtypes)
print('\n💡 Keterangan:')
print('  - int64   = bilangan bulat (integer)')
print('  - float64 = bilangan desimal (float)')
print('  - object  = teks / string (categorical)')
```

## 6. Seleksi Kolom dari DataFrame

Sesuai materi: Membuang kolom ID yang tidak bermakna, mulai dari kolom **`player_name`** menggunakan **`df.iloc[:,:]`**

```python
# Pilih semua kolom mulai dari player_name (semua kolom dalam kasus ini sudah bersih)
df_noid = df.copy()

# Alternatif: pilih kolom tertentu saja
kolom_penting = ['player_name', 'team_title', 'games', 'goals', 'assists', 'shots', 'xG', 'yellow_cards', 'red_cards']
df_noid = df[kolom_penting].copy()

print(f'DataFrame setelah seleksi kolom: {df_noid.shape}')
df_noid.head(5)
```

## 7. Menampilkan Data dalam Urutan Tertentu

Sesuai materi: **`sort_values()`** untuk mengurutkan data.

```python
# Sort berdasarkan player_name (ascending = A-Z)
df1 = df_noid.sort_values(by='player_name', ascending=True)
print('=== Sort by player_name (A-Z) — 10 baris pertama ===')
df1.head(10)
```

```python
# Sort berdasarkan assists (descending), jika sama → sort by team_title (ascending)
df1 = df_noid.sort_values(
    by=['assists', 'team_title'],
    ascending=[False, True]
)
print('=== Sort by assists (DESC) lalu team_title (ASC) — 10 baris pertama ===')
df1.head(10)
```

```python
# Sort berdasarkan goals terbanyak
df_top_scorer = df_noid.sort_values(by='goals', ascending=False)
print('=== Top 10 Pencetak Gol Terbanyak ===')
df_top_scorer[['player_name', 'team_title', 'goals', 'assists']].head(10)
```

## 8. Ringkasan

| Fungsi | Kegunaan |
|---|---|
| `pd.read_csv(path)` | Import file CSV ke DataFrame |
| `df.head(n)` | Tampilkan n baris pertama |
| `df.tail(n)` | Tampilkan n baris terakhir |
| `df.dtypes` | Cek tipe data tiap kolom |
| `df.iloc[:, n:]` | Pilih kolom dari index ke-n |
| `df[['col1','col2']]` | Pilih kolom tertentu |
| `df.sort_values(by=...)` | Urutkan data berdasarkan kolom |
