## 0. Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset dari Notebook 1
path = 'epl_goalscorer_2021.csv'
df = pd.read_csv(path)

# Buat df_noid (tanpa kolom teks yang tidak perlu untuk statistik)
df_noid = df.copy()

print('✅ Dataset dimuat!')
print(f'Shape: {df_noid.shape}')
df_noid.head(3)
```

## 2. Central Tendency — Arithmetic Mean (Rata-rata)

**Definisi:** Jumlah semua nilai dibagi jumlah data.

$$\bar{x} = \frac{1}{N} \sum_{i=1}^{N} x_i$$

Sesuai materi: gunakan **`mean()`** di Pandas.

```python
# Mean goals, games, assists, shots
mean_result = df_noid[['goals', 'games', 'assists', 'shots']].mean()
print('=== Rata-rata (Mean) ===')
print(mean_result)
print()
print(f'🔍 Interpretasi:')
print(f'  Rata-rata gol per pemain  : {mean_result["goals"]:.2f}')
print(f'  Rata-rata pertandingan    : {mean_result["games"]:.2f}')
print(f'  Rata-rata assist          : {mean_result["assists"]:.2f}')
print(f'  Rata-rata shots           : {mean_result["shots"]:.2f}')
```

```python
# Verifikasi manual untuk goals
total_goals = df_noid['goals'].sum()
n_players = len(df_noid)
mean_manual = total_goals / n_players
print(f'Total gol semua pemain : {total_goals}')
print(f'Jumlah pemain          : {n_players}')
print(f'Mean manual (total/n)  : {mean_manual:.4f}')
print(f'Mean via pandas        : {df_noid["goals"].mean():.4f}')
print(f'✅ Hasil sama!' if abs(mean_manual - df_noid['goals'].mean()) < 0.0001 else '❌ Berbeda!')
```

---
## 3. Central Tendency — Median

**Definisi:** Nilai tengah setelah data diurutkan. Lebih tahan terhadap outlier dibanding mean.

- Jika N **ganjil** → nilai di posisi (N+1)/2
- Jika N **genap** → rata-rata dua nilai tengah

Sesuai materi: gunakan **`median()`**

```python
# Median goals, games, assists, shots
median_result = df_noid[['goals', 'games', 'assists', 'shots']].median()
print('=== Median ===')
print(median_result)
```

```python
# Perbandingan Mean vs Median (penting untuk cek skewness distribusi)
print('=== Perbandingan Mean vs Median ===')
comparison = pd.DataFrame({
    'Mean'  : df_noid[['goals','games','assists','shots']].mean(),
    'Median': df_noid[['goals','games','assists','shots']].median(),
})
comparison['Mean > Median?'] = comparison['Mean'] > comparison['Median']
comparison['Keterangan'] = comparison['Mean > Median?'].map(
    {True: '📈 Right-skewed (positif)', False: '📉 Left-skewed / Simetris'}
)
print(comparison)
```

---
## 4. Central Tendency — Mode

**Definisi:** Nilai yang paling sering muncul. Bisa lebih dari satu.

Sesuai materi: Digunakan untuk data **nominal/kategorikal**. Gunakan **`mode()`**

```python
# Mode dari kolom team_title (tim yang paling banyak punya pemain)
print('=== Mode team_title (Tim dengan pemain terbanyak) ===')
mode_team = df_noid[['team_title']].mode()
print(mode_team)
print()
print('=== Frekuensi tiap tim (value_counts) ===')
print(df_noid['team_title'].value_counts().head(10))
```

```python
# Mode untuk data numerik (goals)
mode_goals = df_noid[['goals']].mode()
print('=== Mode goals (nilai gol yang paling sering muncul) ===')
print(mode_goals)
print()
print('Distribusi nilai goals:')
print(df_noid['goals'].value_counts().sort_index())
```

---
## 5. Distribution — Range

**Definisi:** Selisih antara nilai maksimum dan minimum.

> ⚠️ Range sensitif terhadap outlier karena hanya bergantung pada 2 nilai ekstrim.

```python
# Tidak ada fungsi range() khusus di Pandas, gunakan max() - min()
cols = ['goals', 'games', 'assists', 'shots']
print('=== Range (Max - Min) ===')
for col in cols:
    min_val = df_noid[col].min()
    max_val = df_noid[col].max()
    range_val = max_val - min_val
    print(f'{col:10s}: min={min_val:4}, max={max_val:4}, range={range_val:4}')
```

---
## 6. Distribution — Quantiles & Quartiles

**Definisi:**
- Q1 (25%) → 25% data berada di bawah nilai ini
- Q2 / Median (50%) → nilai tengah
- Q3 (75%) → 75% data berada di bawah nilai ini

Sesuai materi: gunakan **`quantile(q=...)`**

```python
# Q3 dari semua kolom numerik
print('=== Kuartil ke-3 (Q3 = 75%) semua kolom numerik ===')
df_noid.select_dtypes(include='number').quantile(0.75)
```

```python
# Q3 dari kolom goals saja
q3_goals = df_noid[['goals']].quantile(q=0.75)
print('=== Q3 goals ===')
print(q3_goals)

print()
print('=== Semua Quartile untuk goals ===')
for q_label, q_val in [('Q1 (25%)', 0.25), ('Q2/Median (50%)', 0.50), ('Q3 (75%)', 0.75)]:
    val = df_noid['goals'].quantile(q=q_val)
    print(f'  {q_label}: {val}')
```

```python
# Tabel lengkap Q1, Q2, Q3 untuk semua kolom numerik
numeric_cols = df_noid.select_dtypes(include='number').columns
quartile_df = pd.DataFrame({
    'Q1 (25%)': df_noid[numeric_cols].quantile(0.25),
    'Q2/Median': df_noid[numeric_cols].quantile(0.50),
    'Q3 (75%)': df_noid[numeric_cols].quantile(0.75),
    'IQR (Q3-Q1)': df_noid[numeric_cols].quantile(0.75) - df_noid[numeric_cols].quantile(0.25),
})
print('=== Rangkuman Quartile ===')
quartile_df
```

## 7. Distribution — Standard Deviation & Variance

**Definisi:**
$$s = \sqrt{\frac{\sum_{i=1}^{N}(x_i - \bar{x})^2}{N-1}}$$

- **Std Dev besar** → data menyebar jauh dari mean
- **Std Dev kecil** → data berkumpul dekat mean
- **Variance** = std dev²

```python
# Variance dan Std Dev untuk goals
var_goals = df_noid[['goals']].var()
std_goals = df_noid[['goals']].std()

print(f'=== Variance dan Std Dev untuk goals ===')
print(f'Variance  : {var_goals["goals"]:.4f}')
print(f'Std Dev   : {std_goals["goals"]:.4f}')
print(f'Mean      : {df_noid["goals"].mean():.4f}')
print()
print(f'💡 Kebanyakan pemain mencetak gol dalam rentang:')
mean_g = df_noid['goals'].mean()
std_g = df_noid['goals'].std()
print(f'   [{mean_g - std_g:.1f}, {mean_g + std_g:.1f}] (±1σ)')
```

```python
# Ringkasan statistik lengkap untuk goals
print('=== Ringkasan Lengkap Distribusi Goals ===')
stats_goals = {
    'Count'   : df_noid['goals'].count(),
    'Mean'    : df_noid['goals'].mean(),
    'Median'  : df_noid['goals'].median(),
    'Std Dev' : df_noid['goals'].std(),
    'Variance': df_noid['goals'].var(),
    'Min'     : df_noid['goals'].min(),
    'Max'     : df_noid['goals'].max(),
    'Range'   : df_noid['goals'].max() - df_noid['goals'].min(),
    'Q1'      : df_noid['goals'].quantile(0.25),
    'Q3'      : df_noid['goals'].quantile(0.75),
    'IQR'     : df_noid['goals'].quantile(0.75) - df_noid['goals'].quantile(0.25),
}
for k, v in stats_goals.items():
    print(f'  {k:10s}: {v:.4f}')
```

## 8. Outliers — Metode 3-Sigma Rule

Sesuai materi: data disebut **outlier** jika jaraknya lebih dari 3× std dev dari mean.

- Outlier jika: $x_i < \bar{x} - 3\sigma$ atau $x_i > \bar{x} + 3\sigma$
- Probabilitas: hanya **0.27%** data yang wajar berada di luar ±3σ

```python
# Deteksi outlier dengan 3-Sigma Rule pada kolom goals
mean_g = df_noid['goals'].mean()
std_g  = df_noid['goals'].std()

iso = (df_noid['goals'] < mean_g - 3*std_g) | (df_noid['goals'] > mean_g + 3*std_g)

df1 = df_noid[['player_name', 'goals']].assign(is_outlier=iso)

print(f'Mean  : {mean_g:.4f}')
print(f'StdDev: {std_g:.4f}')
print(f'Batas bawah  (μ - 3σ): {mean_g - 3*std_g:.2f}')
print(f'Batas atas   (μ + 3σ): {mean_g + 3*std_g:.2f}')
print()
print('=== Pemain yang teridentifikasi sebagai Outlier (3-Sigma) ===')
print(df1.loc[df1['is_outlier']])
print(f'\nTotal outlier: {iso.sum()} dari {len(df_noid)} pemain')
```

---
## 9. Outliers — Tukey's Fences (IQR Method)

Sesuai materi:
- **Outlier biasa** jika: $x_i < Q1 - 1.5 \times IQR$ atau $x_i > Q3 + 1.5 \times IQR$
- **Extreme outlier** jika: $x_i < Q1 - 3 \times IQR$ atau $x_i > Q3 + 3 \times IQR$

```python
# Tukey's Fences untuk goals
q1  = df_noid['goals'].quantile(q=0.25)
q3  = df_noid['goals'].quantile(q=0.75)
iqr = q3 - q1

iso  = (df_noid['goals'] < q1 - 1.5*iqr) | (df_noid['goals'] > q3 + 1.5*iqr)
iseo = (df_noid['goals'] < q1 - 3*iqr)   | (df_noid['goals'] > q3 + 3*iqr)

df1 = df_noid[['player_name', 'goals']].assign(
    is_outlier=iso,
    is_extreme_outlier=iseo
)

print(f'Q1  : {q1}')
print(f'Q3  : {q3}')
print(f'IQR : {iqr}')
print(f'Batas outlier bawah  (Q1 - 1.5×IQR): {q1 - 1.5*iqr:.2f}')
print(f'Batas outlier atas   (Q3 + 1.5×IQR): {q3 + 1.5*iqr:.2f}')
print(f'Batas extreme bawah  (Q1 - 3×IQR)  : {q1 - 3*iqr:.2f}')
print(f'Batas extreme atas   (Q3 + 3×IQR)  : {q3 + 3*iqr:.2f}')
print()
print('=== Outlier & Extreme Outlier (Tukey) ===')
df1.loc[df1['is_outlier'] | df1['is_extreme_outlier']]
```

## 10. Frequency Table & Grouping

Sesuai materi: **`value_counts()`** untuk melihat frekuensi data nominal.

```python
# Frekuensi pemain per tim (value_counts)
print('=== Jumlah Pemain per Tim (value_counts) ===')
print(df_noid['team_title'].value_counts())
print()
print('💡 Nilai tertinggi = Mode dari kolom team_title')
```
